## Seaborn의 통계 차트 및 데이타 시각화 예제

- https://teddylee777.github.io/visualization/seaborn-tutorial-1

- seaborn은 matplotlib의 상위 호환 데이터 시각화를 위한 라이브러리입니다.
seaborn패키지는 데이터프레임으로 다양한 통계 지표를 낼 수 있는 시각화 차트를 제공하기 때문에 데이터 분석에 활발히 사용되고 있는 라이브러리입니다.


- 이번 seaborn 튜토리얼에서는 matplotlib에는 없고 seaborn에서만 제공되는 통계 차트 위주로 설명드리겠습니다.

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager
```

```python
# 한글 폰트 적용
font_fname = './NanumBarunGothic.ttf'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
plt.rc('font', family=font_name)

# 캔버스 사이즈 적용
plt.rcParams['figure.figsize'] = (12,9)
```

### 통계 기반의 시각화를 제공해주는 Seaborn

- seaborn 라이브러리가 매력적인 이유는 바로 "통계 차트" 이다

- https://seaborn.pydata.org/api.html

```python
# sample data 가져오기

titanic = sns.load_dataset('titanic')
titanic.head()
```

- survived: 생존여부
- pclass: 좌석등급
- sex: 성별
- age: 나이
- sibsp: 형제자매 숫자
- parch: 부모자식 숫자
- fare: 요금
- embarked: 탑승 항구
- class: 좌석등급 (영문)
- who: 사람 구분
- deck: 데크
- embark_town: 탑승 항구 (영문)
- alive: 생존여부 (영문)
- alone: 혼자인지 여부

```python
# sample data 가져오기

tips = sns.load_dataset('tips')
tips.tail()
```

- total_bill: 총 합계 요금표
- tip: 팁
- sex: 성별
- smoker: 흡연자 여부
- day: 요일
- time: 식사 시간
- size: 식사 인원

### 1. Countplot

항목별 갯수를 세어주는 plot.

알아서 해당 column을 구성하고 있는 value들을 구분하여 보여준다.

- https://seaborn.pydata.org/generated/seaborn.countplot.html

### 1-1 세로 그래프 그리기

```python
sns.countplot(x='class', hue="who", data=titanic)
plt.show()
```

### 1-2 가로 그래프 그리기

```python
sns.countplot(y='class', hue='who', data=titanic)
plt.show()
```

### 1-3 색상 팔레트 설정

```python
sns.countplot(x='class', hue='who', palette='Paired', data=titanic)
plt.show()
```

### 2. distplot

matplotlib의 hist 그래프와 kdeplot를 통합한 그래프

- 분포와 밀도를 확인

- https://seaborn.pydata.org/generated/seaborn.distplot.html?highlight=distplot#seaborn.distplot

```python
# 샘플데이타 생성

x = np.random.random(100)
```

### 2-1 기본 distplot

`distplot`은 더 이상 사용되지 않는 기능이며 향후 버전에서 제거 될 예정입니다.

'displot'(유연성이 유사한 그림 수준 함수) 또는 'histplot'(히스토그램을위한 축 수준 함수)을 사용하도록 코드를 조정하세요.

```python
sns.distplot(x)
plt.rc('axes', unicode_minus=False)
plt.show()
```

### 3. heatmap

색상으로 표현할 수 있는 다양한 정보를 이정한 이미지 위에 열분포 형태의 비쥬얼한 그래픽으로 축력

- https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap

### 3-1 기본 heatmap

```python
# 샘플 데이터

uniform_data = np.random.rand(10,12) # 10 행 x 12 열 
uniform_data[9][11]
```

```python
sns.heatmap(uniform_data, annot=True)
plt.show()
```

### 3-2 pivot table을 활용하여 그리기

```python
# 데이터 확인 
tips.tail(10)
```

```python
# pivot
pivot = tips.pivot_table(index='day', columns='size', values='tip')
pivot
```

```python
sns.heatmap(pivot, annot=True)
plt.show()
```

### 3-3 correlation(상관관계) 시각화

corr() 함수는 데이터의 상관관계를 보여준다.

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(),annot=True)
plt.show()
```

```python
# 색 지정
sns.heatmap(titanic.corr(),annot=True, cmap='YlGnBu')
plt.show()
```

### 4. pairplot

- grid 형태로 각 집합의 조합에 대해 히스토그램과 분포도를 그린다.

- 숫자형 column에 대해서만 그린다.

- https://seaborn.pydata.org/generated/seaborn.pairplot.html?highlight=pairplot#seaborn.pairplot

```python
# data 확인
tips.head()
```

### 4-1 기본 pairplot 그리기

```python
sns.pairplot(tips)
plt.show()
```

### 4-2 hue 옵션으로 특성을 구분

```python
sns.pairplot(tips, hue='size')
plt.show()
```

### 4-3 컬러 팔레트 적용

```python
sns.pairplot(tips, hue='size', palette='rainbow')
plt.show()
```

### 4-4 사이즈 적용

```python
sns.pairplot(tips, hue='size', palette='rainbow', height=5)
plt.show()
```

###  5. violinplot

바이올린처럼 생겨 violinplot

column 에 대한 데이터의 비교 분포도를 확인할 수 있다.
- 곡선진 부분(뚱뚱한 부분)은 데이터의 분포를 나타낸다.
- 양쪽 끝 뾰족한 부분은 데이터의 최소값과 최대값을 나타낸다.

- https://seaborn.pydata.org/generated/seaborn.violinplot.html

### 5-1 기본 violinplot 그리기

```python
sns.violinplot(x=tips['total_bill'])
plt.show()
```

### 5-2 비교 분포 확인

x, y 축을 지정해줌으로써 바이올린을 분할하여 비교분포"를 볼 수 있다.

```python
sns.violinplot(x='day',y='total_bill', data=tips)
plt.show()
```

### 5-3 가로로 누운 violinplot

```python
sns.violinplot(y='day', x='total_bill', data=tips)
plt.show()
```

### 5-4 hue 옵션으로 분포 비교

사실 hue 옵션을 사용하지 않으면 바이올린이 대칭되기 때문에 비교 분포의 큰 의미가 없다.
하지만, hue 옵션을 주면, "단일 column에 대한 바이올린 모양의 비교"를 할 수 있다.

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted')
plt.show()
```

### 5-5 split 옵션으로 바이올린을 합쳐서 볼 수 있다.

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted', split=True)
plt.show()
```

### 6. lmplot

lmplot은 column 간의 선형관계를 확인하기에 용이한 챠트

또한, outlier도 같이 짐작해 볼 수 있다.

- https://seaborn.pydata.org/generated/seaborn.lmplot.html

### 6-1 기본 lmplot

```python
sns.lmplot(x='total_bill', y='tip', height=8, data=tips)
plt.show()
```

### 6-2 hue 옵션으로 다중 선형관계 그리기

```python
# 아래 그래프를 통하여 비흡연자가 흡연자 대비 좀 더 가파를 선형관계를 가지는 것을 볼 수 있다.

sns.lmplot(x='total_bill', y='tip', hue='smoker', height=8, data=tips)
plt.show()
```

### 6-3 col 옵션을 추가하여 그래프를 별도로 그려볼 수 있다.

```python
# col_wrap으로 한줄에 표기할 column의 갯수를 명시할 수 있다.

sns.lmplot(x='total_bill', y='tip', hue='smoker', col='day', col_wrap=2, height=8, data=tips)
plt.show()
```

### 7 relplot

두 column간 상관관계를 보지만, lmplot처럼 선형관계를 따로 그리지는 않는다.

- https://seaborn.pydata.org/generated/seaborn.relplot.html?highlight=relplot#seaborn.relplot

### 7-1 기본 relplot

```python
sns.relplot(x='total_bill', y='tip', hue='day', data=tips)
plt.show()
```

### 7-2 col 옵션으로 그래프 분할

```python
sns.relplot(x='total_bill', y='tip', hue='day', col='time', data=tips)
plt.show()
```

### 7-3 row와 column에 표기할 데이타 column 선택

```python
sns.relplot(x='total_bill', y='tip', hue='day', row='sex', col='time', data=tips)
plt.show()
```

### 7-4 컬러 팔레트 적용

```python
sns.relplot(x='total_bill', y='tip', hue='day', row='sex', col='time', palette='CMRmap_r', data=tips)
plt.show()
```

### 8. jointplot

scatter(산점도)와 histogram(분포)를 동시에 그려준다.

숫자형 데이터만 표현 가능.

- https://seaborn.pydata.org/generated/seaborn.jointplot.html?highlight=jointplot#seaborn.jointplot

### 8-1 기본 jointplot 그리기

```python
sns.jointplot(x="total_bill", y="tip", height=8, data=tips)
plt.show()
```

### 8-2 선형관계를 표현하는 regression (회귀) 라인 그리기

```python
# 옵션에 kind='reg' 추가

sns.jointplot(x='total_bill', y='tip', height=8, data=tips, kind='reg')
plt.show()
```

### 8-3 hex 밀도 보기

```python
# kind='hex' 옵션을 통해 hex 모양의 밀도를 확인할 수 있다.

sns.jointplot(x='total_bill', y='tip', height=8, data=tips, kind='hex')
plt.show()
```

### 8-4 등고선 모양으로 밀집도 확인하기

```python
# sample data load

iris = sns.load_dataset('iris')
iris
```

```python
# kind='kde' 옵션으로 데이터의 밀집도를 보다 부드러운 선으로 확인할 수 있다.

sns.jointplot(x='sepal_length', y='petal_length', height=8, data=iris, kind='kde', color='g')
plt.show()
```
