```python
from datetime import datetime

datetime.today()            # 현재 날짜 가져오기

# datetime.today().year        # 현재 연도 가져오기

# datetime.today().month      # 현재 월 가져오기

# datetime.today().day        # 현재 일 가져오기

# datetime.today().hour        # 현재 시간 가져오기

# datetime.today().weekday()    # 요일 가져오기
```

```python
# 유튜브 동영상 강의 사용 일자 
date = '20200507'

# 오늘 날자 스트링 만들기

year = repr(datetime.today().year)        # 현재 연도 가져오기

month = repr(datetime.today().month)      # 현재 월 가져오기
if len(month) < 2 : month = '0' + month

day = repr(datetime.today().day)          # 현재 일 가져오기
# 주말(토, 일)이면, 금요일로 날짜 설정
weekday = datetime.today().weekday()
if weekday == 5 :
    # print("토요일")
    day = int(day) - 1
    day = repr(day)
elif weekday == 6 :
    # print("일요일")
    day = int(day) - 2
    day = repr(day)

if len(day) < 2 : day = '0' + day
#
# date = year + month + day

date
```

```python
### 일자별 순매수 Url 에 날짜를 넣는 코드 작성할 것


page=1
# 일자별 순매수
# url = 'https://finance.naver.com/sise/investorDealTrendDay.nhn?bizdate={date}&sosok=&page={page}'.format(date=date,page=page)
# 시간별 순매수
url = 'https://finance.naver.com/sise/investorDealTrendTime.nhn?bizdate=20210312&sosok=&page={page}'.format(page=page)

# 웹페이지 가져오기
resp = requests.get(url)

# 동영상 강의에서 사용한 함수인데, BytesIO()를 쓰지않고,
# request.get에서 가져온 데이타를 아래와 같이 사용해도 된다.
# html = BytesIO(resp.content)
html = resp.text # 또는 html = resp.content

html
url
```

```python
# html 코드에서 table(표)을 읽어온다
df = pd.read_html(html)[0]
df

# requests.get(url) 해서 가져오지 않고,
# pandas에서 바로 가져와도 된다.
# pd.read_html(url)[0]
```

```python
# html 코드에서 table(표)을 읽어온다, 두번째 Table
pd.read_html(html)[1]
```

```python
# 마지막 페이지 찾기
# 두번째 Table을 이용해서, 마지막 읽어올 페이지를 찾는다.

soup = bs(html, "html.parser")

str = soup.select('td.pgRR > a')[0]['href']
str = str.split('=')
i = len(str)
end_page = int(str[i-1])

end_page
```

```python
### 칼럼만 있는 데이타 프레임을 만든다.
cols = df.columns
# cols

df2 = pd.DataFrame(columns=cols)
# df2   # column이 두줄로 나온다.

df2.columns = df2.columns.droplevel() # 칼럼 한줄을 지운다.
df2

# 이렇게 만든 칼럼만 있는 dataframe 에 1 ~ 마지막 페이지까지의 데이터를 붙인다.
# 근데, 이렇게 하지않고 붙이고, 다음에 칼럼을 한줄 지운거랑 비교해보자.

### 아래 check 1 에서 보듯이

# 예상 또는 강의에서 말한대로 되지않고 칼럼이 이상하게 나온다.
# 그래서, 이 코드들이 사실은 필요가 없다.
```

```python
# 위에서 마지막 페이지와 칼럼명을 가져오기위해 사용했던, requests를 사용하지 않고,
# 바로 padas의 read_html로 (table) 데이타를 가져오도록 한다.

# 읽어올 페이지의 1 ~ end_page까지의 페이지를 읽어와서 데이타프레임을 아래로 붙인다.
for page in tqdm(range(1,end_page)):
    
    # 일자별 순매수
    url = 'https://finance.naver.com/sise/investorDealTrendDay.nhn?bizdate={date}&sosok=&page={page}'.format(date=date,page=page)
    # 시간별 순매수
#     url = 'https://finance.naver.com/sise/investorDealTrendTime.nhn?bizdate=20210312&sosok=&page={page}'.format(page=page)

    # resp = requests.get(url)
    # html = BytesIO(resp.content)

    df_single = pd.read_html(html)[0]
    df = pd.concat([df, df_single])
    
    # 위에서 칼럼만 있는 데이타 프레임에 데이타프레임을 붙인다.
    df2 = pd.concat([df2, df_single]) # 근데, 칼럼이 좀 이상하다.
    
df.head(20)      # 인덱스를 보면, 14번 행 다음에 다시 0 으로 시작
```

```python
# 결측치 정리

# df.dropna( axis = 1 ) # 결측치가 있는 열 삭제 # 인덱스 빼고 다 지워진다 ㅋ
df = df.dropna( axis = 0 ) # 결측치가 있는 행 삭제

df.head(20)
```

```python
# 인덱스 번호를 다시 정의한다. 
# 각 페이지를 가져와서, 붙이다(caoncat)보니, 인덱스가 연속직이지 않다.

df.reset_index(drop=True, inplace=True)

df.head(20)
```

```python
# 칼럼이 두줄로 되어있어, 위쪽 칼럼을 지운다.
df.columns = df.columns.droplevel()

df.head()
```
