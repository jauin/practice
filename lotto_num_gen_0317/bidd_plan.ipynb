```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
keyword=['나무']#,'재선충','풀']
df = pd.DataFrame()
for key in keyword: 
    url = f'https://contract.ulsan.go.kr/order_schedule/all?instCd=&start_day=&end_day=&construction_name={key}'
    resp = requests.get(url)
    html = bs(resp.content, 'html.parser')
#     html = html.select('table.mat30.table_board')
#     print(len(html))
    html = html.select('body')
    html_str=str(html)
    print(html_str)
#     print(html)
    df = pd.read_html(html_str)
# #     pd.read_html(htm, skiprows=2, flavor='bs4')[0]
    df
```
